# Plot -log (used in Fast R-CNN)

In Fast R-CNN the classification loss in the multi-task-loss is defined by L_cls(p,u):=-log p_u where u is the index of the ground truth class.

```python
import numpy as np

step = 0.01
x = np.arange(step,1.0,step)
#print(x)
y1 = [np.log(val) for val in x]
y2 = [-np.log(val) for val in x]
plt.plot(x,y1, label="log(x)", linestyle="--")
plt.plot(x,y2, label="-log(x)", linestyle=":")
plt.grid()
plt.legend()
plt.show()
```

# Plot smooth(x)

smooth(x) is used in the localization loss in Fast R-CNN.

```python
import numpy as np

def smooth(x):
    if abs(x)<1:
        return 0.5*x**2
    else:
        return abs(x)-0.5

step =0.05
x = np.arange(-5.0,+5.0,step)
#print(x)
y3 = [smooth(val) for val in x]
plt.plot(x,y3, label="smooth(x)", linestyle="-")
plt.grid()
plt.legend()
plt.show()
```

# Exponential function

```python
step = 0.01
x = np.arange(-5.0,5.0,step)
#print(x)
y = [np.exp(val) for val in x]
plt.plot(x,y, label="exp(x)", linestyle="--")
plt.grid()
plt.legend()
plt.show()
```
