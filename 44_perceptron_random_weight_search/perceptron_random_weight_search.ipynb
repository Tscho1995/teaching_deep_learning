# Einleitung

Wir haben die Perzeptron-Lernregel kennengelernt. Könnte man nicht einfach zufällig Gewichtsmatrizen erzeugen, auf den Testdaten testen wie gut diese Gewichte sind und am Ende die beste Gewichtsmatrix nehmen?

# Trainings- und Testdaten vorbereiten

```python
from tensorflow.examples.tutorials.mnist import input_data

# download and read data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)

x_train = mnist.train.images
y_train = mnist.train.labels
x_test = mnist.test.images
y_test = mnist.test.labels
```

# Hilfsfunktionen: Perzeptron erstellen und testen

```python
import numpy as np
from random import randint

# RELU
def f(x):
    if x < 0:
        return 0
    else:
        return x
    
f = np.vectorize(f)

def create_new_perceptron(nr_inputs, nr_outputs):
    low = -1
    high = +1
    W = np.random.uniform(low,high, (nr_inputs, nr_outputs) )
    return W


def test_perceptron(W, x_test, y_test):
    
    nr_test_images = x_test.shape[0]    
    correct = 0
    for test_img_nr in range(0,nr_test_images):
        
        # Berechnet den Output
        row_vec = x_test[test_img_nr, :]
        x = np.append(row_vec,[1])
        out = predict(W,x)
        
        # Was wäre richtig gewesen?        
        one_hot_vec = y_test[test_img_nr, :]
        gt_label = np.argmax( one_hot_vec )

        # Was wurde prädiziert?
        is_label = np.argmax( out )

        #print(gt_label, is_label)
        if is_label == gt_label:
            correct += 1     
            
    print("Korrekt: {} von {}".format(correct,nr_test_images))
    
    return correct/nr_test_images
 

def predict(W,x):
    act = np.matmul(x, W)
    out = f(act)
    return out
```

# Zufällige Perzeptrons erstellen und testen

```python
nr_tries = 50
best_W = None
best_correct_ratio = 0.0

for try_nr in range(0,nr_tries):
    
    # 1. Perzeptron mit zufälliger
    #    Gewichtsmatrix erstellen
    nr_inputs = 28*28 + 1
    nr_outputs = 10
    W=create_new_perceptron(nr_inputs, nr_outputs)
    
    # 2. Das Perzeptron auf den "Trainingsdaten"
    #    testen    
    correct_ratio = test_perceptron(W, x_train, y_train)
    
    # Interessante Variante:
    # Einschätzen der Güte des aktuellen Perzeptrons W
    # nicht auf Basis aller Trainingsdaten (55.000!),
    # sondern nur auf einer kleinen Teilmenge (z.B. 1000)
    # Vorteil: schneller!
    # Nachteil: Einschätzung der Generalisierungsfähigkeit
    #           des Modells bei einer kleinen Teilmenge der
    #           Trainingsdaten ist nicht so zuverlässig!
    #correct_ratio = test_perceptron(W, x_train[:100], y_train[:100])
    
    # 3. Ausgabe, wie gut das Perzeptron auf
    #    den Trainingsdaten ist
    print("Versuch {} von {} -> {:.2f}% (bisher am besten: {:.2f}%)"
          .format(try_nr,
                  nr_tries,
                  correct_ratio*100.0,
                  best_correct_ratio*100.0))
    
    # 4. Besseres Perzeptron gefunden?
    if correct_ratio>best_correct_ratio:
        best_correct_ratio = correct_ratio
        best_W = W
        
print("'Suche' beendet.")

correct_ratio = test_perceptron(best_W, x_test, y_test)
print("\nAuf den Testdaten erreicht das hinsichtlich Trainingsdaten")
print("beste Perzeptron folgende Klassifikationsrate:")
print("{0}".format(correct_ratio*100.0))        
```
